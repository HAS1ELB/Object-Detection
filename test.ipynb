```python
from tensorflow.keras.models import load_model
import cv2
import utils
import numpy as np

# Charger le modèle TensorFlow
model_tf = load_model("Models/tensorflow_model.h5")

# Préparer l'image
image_path = "Images/2.png"
original_image = cv2.imread(image_path)
image = utils.preprocess_image(image_path)

# Prédiction avec le modèle TensorFlow
results = model_tf.predict(image)

# Extraire les prédictions (hypothétique : adaptation selon le modèle)
bboxes = results[0]  # Boîtes englobantes prédictes
classes = np.argmax(results[1], axis=-1)  # Classes prédictes
confidences = np.max(results[1], axis=-1)  # Scores de confiance

# Afficher les résultats dans la console
print("Boîtes englobantes prédictes :", bboxes)
print("Classes prédictes :", classes)
print("Scores de confiance :", confidences)

# Dessiner les prédictions sur une nouvelle image
output_image = original_image.copy()

for bbox, cls, conf in zip(bboxes, classes, confidences):
    x1, y1, x2, y2 = map(int, bbox)
    label = f"{cls} {conf:.2f}"
    color = (0, 255, 0)  # Couleur verte pour les boîtes

    # Dessiner le rectangle autour de l'objet
    cv2.rectangle(output_image, (x1, y1), (x2, y2), color, 2)

    # Ajouter le label au-dessus de la boîte
    cv2.putText(output_image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# Sauvegarder et afficher l'image avec les annotations
output_path = "Images/annotated_image.png"
cv2.imwrite(output_path, output_image)
cv2.imshow("Prédictions Annotées", output_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from tensorflow.keras.models import load_model
from ultralytics import YOLO
import cv2
import utils
import numpy as np

# Charger les modèles
model_tf = load_model("Models/tensorflow_model.h5")
model_yolo = YOLO("Yolo-Weights/yolov8n.pt")

# Préparer l'image
image_path = "Images/1.png"
original_image = cv2.imread(image_path)
image = utils.preprocess_image(image_path)

# Prédiction avec le modèle TensorFlow
results_tf = model_tf.predict(image)

# Extraire les prédictions TensorFlow
bboxes_tf = results_tf[0]  # Boîtes englobantes prédictes
classes_tf = np.argmax(results_tf[1], axis=-1)  # Classes prédictes
confidences_tf = np.max(results_tf[1], axis=-1)  # Scores de confiance

# Prédiction avec YOLO
results_yolo = model_yolo(image_path, stream=True)
bboxes_yolo, classes_yolo, confidences_yolo = [], [], []
for result in results_yolo:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        bboxes_yolo.append([x1, y1, x2, y2])
        classes_yolo.append(int(box.cls[0]))
        confidences_yolo.append(float(box.conf[0]))

# Afficher les résultats TensorFlow dans la console
print("[TensorFlow] Boîtes englobantes :", bboxes_tf)
print("[TensorFlow] Classes :", classes_tf)
print("[TensorFlow] Scores de confiance :", confidences_tf)

# Afficher les résultats YOLO dans la console
print("[YOLO] Boîtes englobantes :", bboxes_yolo)
print("[YOLO] Classes :", classes_yolo)
print("[YOLO] Scores de confiance :", confidences_yolo)

# Dessiner les prédictions sur deux images différentes
output_image_tf = original_image.copy()
output_image_yolo = original_image.copy()

# Dessiner les prédictions TensorFlow
for bbox, cls, conf in zip(bboxes_tf, classes_tf, confidences_tf):
    x1, y1, x2, y2 = map(int, bbox)
    label = f"{cls} {conf:.2f}"
    color = (255, 0, 0)  # Couleur bleue pour TensorFlow
    cv2.rectangle(output_image_tf, (x1, y1), (x2, y2), color, 2)
    cv2.putText(output_image_tf, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# Dessiner les prédictions YOLO
for bbox, cls, conf in zip(bboxes_yolo, classes_yolo, confidences_yolo):
    x1, y1, x2, y2 = bbox
    label = f"{cls} {conf:.2f}"
    color = (0, 255, 0)  # Couleur verte pour YOLO
    cv2.rectangle(output_image_yolo, (x1, y1), (x2, y2), color, 2)
    cv2.putText(output_image_yolo, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# Sauvegarder et afficher les images annotées
output_path_tf = "Images/annotated_image_tf.png"
output_path_yolo = "Images/annotated_image_yolo.png"
cv2.imwrite(output_path_tf, output_image_tf)
cv2.imwrite(output_path_yolo, output_image_yolo)

cv2.imshow("Prédictions TensorFlow", output_image_tf)
cv2.imshow("Prédictions YOLO", output_image_yolo)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Comparaison visuelle des résultats sauvegardés
print(f"Images sauvegardées : {output_path_tf}, {output_path_yolo}")
```

```python
from ultralytics import YOLO
import cv2
import cvzone
import math
import time

# Chargement de la vidéo d'entrée
cap = cv2.VideoCapture("videos/motorbikes.mp4")  # Pour une vidéo
# cap = cv2.VideoCapture(0)  # Pour la webcam
# cap.set(3, 1280)
# cap.set(4, 720)

# Initialisation du modèle YOLO
model = YOLO("Yolo-Weights/yolov8n.pt")

# Liste des classes
classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
              "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
              "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
              "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
              "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
              "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
              "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
              "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
              "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
              "teddy bear", "hair drier", "toothbrush"
              ]

# Initialisation des paramètres pour l'enregistrement vidéo
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec vidéo (mp4v pour .mp4)
frame_width = int(cap.get(3))  # Largeur des frames
frame_height = int(cap.get(4))  # Hauteur des frames
fps = int(cap.get(cv2.CAP_PROP_FPS))  # Fréquence d'images

# Création de l'objet VideoWriter
out = cv2.VideoWriter('output_video.mp4', fourcc, fps, (frame_width, frame_height))

# Variables pour le calcul des FPS
prev_frame_time = 0
new_frame_time = 0

while True:
    new_frame_time = time.time()
    success, img = cap.read()

    # Arrêter la boucle si la vidéo est terminée
    if not success:
        break

    # Détection avec YOLO
    results = model(img, stream=True)
    for r in results:
        boxes = r.boxes
        for box in boxes:
            # Bounding Box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
            # cv2.rectangle(img,(x1,y1),(x2,y2),(255,0,255),3)
            w, h = x2 - x1, y2 - y1
            cvzone.cornerRect(img, (x1, y1, w, h))
            # Confiance
            conf = math.ceil((box.conf[0] * 100)) / 100
            # Nom de la classe
            cls = int(box.cls[0])

            cvzone.putTextRect(img, f'{classNames[cls]} {conf}', (max(0, x1), max(35, y1)), scale=1, thickness=1)

    # Calcul des FPS
    fps = 1 / (new_frame_time - prev_frame_time)
    prev_frame_time = new_frame_time
    print(f"FPS: {fps:.2f}")

    # Affichage de l'image
    cv2.imshow("Image", img)

    # Écriture de l'image modifiée dans le fichier vidéo
    out.write(img)

    # Quitter si 'q' est pressé
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libération des ressources
cap.release()
out.release()
cv2.destroyAllWindows()
```
